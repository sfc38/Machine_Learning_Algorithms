# **Log Likelihood Derivation**


## **Log Likelihood**

The log likelihood for a single observation is given by:

$$l(y\mid x, \theta) = y.ln(g(x,\theta)) +  (1-y).ln(1-g(x,\theta))$$


## **Derivation of Log Likelihood**

The derivative of the log likelihood function with respect to a single parameter $\theta_j$:

$$
\frac{\partial}{\partial \theta_j} l(y\mid x, \theta) 
= \frac{\partial}{\partial \theta_j}y.ln(g(x,\theta))
+ \frac{\partial}{\partial \theta_j}(1-y).ln(1-g(x,\theta))
$$

Before proceeding, let's recall the derivative of the natural logarithm function:

$$
\frac{\partial}{\partial x} (ln x) = \frac{1}{x} . \frac{\partial}{\partial x} (x)
$$

Using this we can now write the derivative of the log likelihood function with respect to $\theta_j$.

$$
\frac{\partial}{\partial \theta_j} l(y\mid x, \theta) 
= \frac{y}{g(x,\theta)} \frac{\partial}{\partial \theta_j} g(x,\theta)
+ \frac{(1-y)}{1-g(x,\theta)} \frac{\partial}{\partial \theta_j} (1-g(x,\theta))
$$

Recalling the sigmoid function:

$$g(x,\theta)= \frac{1}{1+e^{-x\theta}}$$

We can find its derivative:

$$
\frac{\partial}{\partial \theta_j} g(x,\theta) = 
\frac{\partial}{\partial \theta_j} (\frac{1}{1+e^{-x\theta}})
$$

Simplifying further, we have:

$$
\frac{\partial}{\partial \theta_j} g(x,\theta) = 
\frac{\partial}{\partial \theta_j} ({1+e^{-x\theta}})^{-1}
$$

Applying the chain rule, we obtain:

$$
\frac{\partial}{\partial \theta_j} ({1+e^{-x\theta}})^{-1} = 
-1 ({1+e^{-x\theta}})^{-2} (0-x_je^{-x\theta})
$$

Simplifying further, we can write this as:

$$
\frac{\partial}{\partial \theta_j} ({1+e^{-x\theta}})^{-1} = 
\frac{x_je^{-x\theta}}{({1+e^{-x\theta}})^{2}}
$$

Expanding the expression, we have:

$$
\frac{\partial}{\partial \theta_j} ({1+e^{-x\theta}})^{-1} = 
x_j
\frac{1}{{1+e^{-x\theta}}}
\frac{e^{-x\theta}}{{1+e^{-x\theta}}}
$$

Simplifying further, we can rewrite this as:

$$
\frac{\partial}{\partial \theta_j} ({1+e^{-x\theta}})^{-1} = 
x_j
g(x,\theta)
(1-g(x,\theta))
$$

So finally we can write:

$$
\frac{\partial}{\partial \theta_j} g(x,\theta) = 
x_j g(x,\theta) (1-g(x,\theta))
$$

And the derivative of the complementary term:

$$
\frac{\partial}{\partial \theta_j} (1-g(x,\theta)) = 
-x_j g(x,\theta) (1-g(x,\theta))
$$

Now, let's substitute these derivatives into the derivative of the log likelihood function with respect to $\theta_j$:

$$
\frac{\partial}{\partial \theta_j} l(y\mid x, \theta) 
= \frac{y}{g(x,\theta)} \frac{\partial}{\partial \theta_j} g(x,\theta)
+ \frac{(1-y)}{1-g(x,\theta)} \frac{\partial}{\partial \theta_j} (1-g(x,\theta))
$$

$$
= \frac{y}{g(x,\theta)} x_j g(x,\theta) (1-g(x,\theta))
- \frac{(1-y)}{1-g(x,\theta)} x_j g(x,\theta) (1-g(x,\theta))
$$

$$
= (\frac{y}{g(x,\theta)}
- \frac{(1-y)}{1-g(x,\theta)}) x_j g(x,\theta) (1-g(x,\theta))
$$

Simplifying further:

$$
= (\frac{y (1-g(x,\theta)) }{g(x,\theta) (1-g(x,\theta))}
- \frac{(1-y) g(x,\theta)}{(1-g(x,\theta)) g(x,\theta)}) x_j g(x,\theta) (1-g(x,\theta))
$$

$$
= (\frac{y (1-g(x,\theta)) - (1-y) g(x,\theta)}{g(x,\theta) (1-g(x,\theta))} )x_j g(x,\theta) (1-g(x,\theta))
$$

$$
= (y (1-g(x,\theta)) - (1-y) g(x,\theta) )x_j
$$

Simplifying further:

$$
=(y- y g(x,\theta) - g(x,\theta)+y g(x,\theta) ) x_j
$$

$$
=(y - g(x,\theta) ) x_j
$$

Finally we get for derivative of log likelihood:

$$
\frac{\partial}{\partial \theta_j} l(y\mid x, \theta) 
=(y - g(x,\theta) ) x_j
$$
